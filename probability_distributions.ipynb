#**Scipy.stats**

Módulo que contém um grande número de distribuições (contínuas e discretas) de probabilidade. 

Em geral, fornece:

*   distribuições de **densidade de probabilidade** (**pdf** ou **pmf**, probability density function), $P(X=x)$: probabilidade associada a cada valor possível de uma variável aleatória;
*   distribuições **cumulativas** (**cdf**, cumulative density function), $P(X \le x_n) = \sum_{i=0}^{n}P(X=x_i)$: 

Além de funções para calcular varíaveis estatísticas como a média, variância, etc...

https://docs.scipy.org/doc/scipy/reference/stats.html

# **Distribuição binomial:**

$f(n|N,p) = \frac{N!}{n!(N-n)!}p^{n}(1 - p)^{(N-n)}; $


$f(n|N,p) = \left(\begin{array}{c} N \\ n \end{array}\right) p^{n}(1 - p)^{(N-n)}$


A distribuição binomial tem N e p como parâmetros. 

**Média:**

$\mu = E[n] = \sum_{n=0}^{N}nf(n|N, p) = Np$

**Variância:**

$V = E[n^2] - (E[n])^{2} = Np(1 - p)$




```
from scipy.stats import binom  # importar a biblioteca binom de scipy.stats
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html

import matplotlib.pyplot as plt # importar a biblioteca de visualização de dados do matplotlib, pyplot
import numpy as np

def plot_binomial(N, prop, x_textpos, y_textpos):
    "função para plotar distribuições binomais"
    x = np.arange(0,N+1); # arange é uma função do numpy que retorna valores igualmente espaçados no intervalo determinado: https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.arange.html
    data_binom = binom.pmf(x, n=N,p=prop)
    mean, var = binom.stats(N, prop, moments='mv')
    print("média = ", mean, ", variância = ", var)
    plt.stem(data_binom) # plotar linhas verticais em relação à base: https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/stem_plot.html
    plt.xlabel('n')
    plt.ylabel('$f(n|N,p)$')
    plt.text(x_textpos, y_textpos, 'N=%s, p=%s'% (N, prop))
    plt.show()
    
plot_binomial(5, 0.5, 4, 0.25)
plot_binomial(10, 0.5, 7, 0.20)
plot_binomial(20, 0.5, 15, 0.15)
plot_binomial(20, 0.1, 15, 0.25)
plot_binomial(20, 0.2, 15, 0.15)
plot_binomial(20, 0.3, 12, 0.15)

```

# Distribuição de probabilidade de Poisson

$f(n|\nu) = \frac{\nu^{n}}{n!}e^{-\nu}$

A distribuição de Poisson é uma distribuição discreta. Tem $\nu$ como parâmetro.

**média:** $\mu = E[n] = \nu$

**variância:** $V = E[n^{2}] - (E[n])^{2} = \nu$

No **scipy.stats** a distribuição de Poisson é chamada de *poisson*

```
from scipy.stats import poisson  # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

import matplotlib.pyplot as plt 
import numpy as np


def plot_poisson(nu, x_range, x_textpos, y_textpos):
    "função para plotar distribuições de poisson"
    x = np.arange(0,x_range); 
    data_poisson = poisson.pmf(x, mu=nu)
    mean, var = poisson.stats(nu, moments='mv')
    print("média = ", mean, ", variância = ", var)
    plt.stem(x,data_poisson)
    plt.xlabel('n')
    plt.ylabel(r'$f(n|\nu)$')
    plt.text(x_textpos, y_textpos, r'$\nu$= %s'% nu)
    plt.show()
    
plot_poisson(2, 21, 10, 0.2)
plot_poisson(5, 21, 10, 0.1)
plot_poisson(10, 21, 15, 0.1)
plot_poisson(50, 100, 80, 0.05)
```

```
from scipy.stats import poisson, norm  # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

import matplotlib.pyplot as plt 
import numpy as np

x = np.arange(0,100)
nu=50
data_poisson = poisson.pmf(x, mu=nu)
gauss_function = norm.pdf(x,nu,7)
#print(gauss_function)
mean, var = poisson.stats(nu, moments='mv')
print("média = ", mean, ", variância = ", var)
plt.stem(x,data_poisson)
plt.plot(x, gauss_function, color='green', label='gauss')

plt.xlabel('n')
plt.ylabel(r'$f(n|\nu)$')
plt.text(80, 0.05, r'$\nu$= %s'% nu)
plt.show()
```

# Distribuição de $\chi^{2}$

$f(z|n) = \frac{1}{2^{n/2}\Gamma(n/2)}z^{(n/2 -1)}e^{-z/2}$

Tem n (número de graus de liberdade) como parâmetro. Distribuição contínua.



*   **Média:** $\mu = E[z] = n$
*   **Variância:** $V = E[z^{2}] - (E[z])^2 = 2n$


No **scipy.stats** a distribuição de $\chi^2$ é chamada de **chi2**

```
from scipy.stats import chi2  # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

import matplotlib.pyplot as plt 
import numpy as np

x_range=20
n=[1, 2, 5, 10]
c=['red', 'blue', 'green', 'magenta']
x = np.linspace(0,x_range+1, 200)
fig, ax = plt.subplots()

for i in range(len(n)):
    data_chi2 = chi2.pdf(x, df=n[i])
    mean, var = chi2.stats(n[i], moments='mv')
    print("n = ", n[i], ", média = ", mean, ", variância = ", var)
    ax.plot(x,data_chi2, label='n=%s'%n[i], color='%s'%c[i])

legend=ax.legend()
plt.show()
```
